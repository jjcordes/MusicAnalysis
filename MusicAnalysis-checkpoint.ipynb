```python
Tori = set(music5[music5.Artist=='Tori Amos']['Album'].drop_duplicates())
Tori = {x.lower() for x in Tori}
print(Tori)
print("")

AllOwnedWords = []
for album in Tori:
    for word in album.split():
        AllOwnedWords.append(word)

AllOwnedWords = set(AllOwnedWords)
print(AllOwnedWords)
```

```python
AllTori = set(["Little Earthquakes",
"Under the Pink",
"Boys for Pele",
"From the Choirgirl Hotel",
"To Venus and Back",
"Strange Little Girls",
"Scarlet's Walk",
"Welcome to Sunny Florida",
"The Beekeeper",
"American Doll Posse",
"Abnormally Attracted to Sin",
"Midwinter Graces",
"Night of Hunters",
"Gold Dust",
"Unrepentant Geraldines"]
)
AllTori = set(AllTori)
AllToriWords = []

for album in AllTori:
    for word in album.split():
        AllToriWords.append(word.lower())

AllToriWords = set(AllToriWords)

Missing = AllToriWords - AllOwnedWords
print("Possibly missing: ", Missing)


GetEm = []
for album in AllTori:
    for word in Missing:
        if album.lower().startswith(word) or album.lower().endswith(word):
            if album not in GetEm:
                GetEm.append(album)
            
print(GetEm)
```

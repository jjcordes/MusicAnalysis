```python
import random

total_heads = 0
total_tails = 0

experiment = 0

headswin = 0
tailswin = 0

while experiment < 10000:
    count = 0
    stringofcoins = ""
    lastflip=""

    head_after_head = 0
    tail_after_head = 0
    
    while count < 10:
        coin = random.randint(1, 2)

        if coin == 1:
            if lastflip == "H":
                head_after_head +=1

            #stringofcoins += "H"
            total_heads += 1

            lastflip = "H"

        elif coin == 2:
            if lastflip == "H":
                tail_after_head +=1

            #stringofcoins += "T"
            total_tails += 1

            lastflip = "T" 

        count += 1

    #print (stringofcoins)
    print ("Heads: " + str(total_heads))
    print ("Tails: " + str(total_tails))
    print ("\n")

    print ("Heads After Heads: " + str(head_after_head))
    print ("Tails After Heads: " + str(tail_after_head))
    print ("\n")
    print ("\n")

    if head_after_head > tail_after_head:
        headswin += 1
    elif tail_after_head > head_after_head:
        tailswin += 1
    
    experiment += 1
    
print("heads won: " + str(headswin))
print("tails won: " + str(tailswin))
```
